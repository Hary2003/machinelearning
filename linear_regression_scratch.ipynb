```python
import pandas as pd

df = pd.read_csv(r"C:\Users\haris\OneDrive\Documents\students_mark.csv")

# Display first few rows
df.head()
```

```python
def loss_function(m, b, points):
    total_error = 0
    for i in range(len(points)):
        x = points.iloc[i].Hours
        y = points.iloc[i].Marks
        total_error += ( y -(m * x + b))**2
    return total_error / float(len(points))
```

```python
def gradient_descent ( points, m_now,b_now,L):
    m_gradient = 0
    b_gradient = 0
    n = len(points)
    for i in range(n):
        x = points.iloc[i].Hours
        y = points.iloc[i].Marks

        m_gradient += -2/n * x * (y -(m_now * x + b_now))
        b_gradient += -2/n * (y -(m_now * x + b_now))

    m = m_now - (L * m_gradient)
    b = b_now - (L * b_gradient)

    return m,b
```

```python
m =0
b = 0
L = 0.0001
epochs = 300

for i in range(epochs):
    m,b = gradient_descent(df,m,b,L)

print(m,b)
```

```python
import matplotlib.pyplot as plt

# Scatter plot of actual data
plt.scatter(df['Hours'], df['Marks'], color='blue', label='Actual Data')

# Regression line
predicted_marks = m * df['Hours'] + b
plt.plot(df['Hours'], predicted_marks, color='red', label='Regression Line')

# Labels and title
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')
plt.title('Linear Regression using Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```
